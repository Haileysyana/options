```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as si
S,T,r,sigma=90,0.5,0.01,0.3
K_value=np.arange(50.0,150.0,5.0)
def BS(S,K,T,r,sigma):
    d1=(np.log(S/K)+(r+0.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d2=(np.log(S/K)+(r-0.5*sigma**2)*T)/(sigma*np.sqrt(T))
    call = S*si.norm.cdf(d1,0.0,1.0)-K*np.exp(-r*T)*si.norm.cdf(d2,0.0,1.0)
    put = -S*si.norm.cdf(-d1,0.0,1.0)+K*np.exp(-r*T)*si.norm.cdf(-d2,0.0,1.0)
    return call,put
K_range = np.linspace(10, 150, 40)
call_prices = []
put_prices = []
for K in K_range:
    call, put = BS(S, K, T, r, sigma)
    call_prices.append(call)
    put_prices.append(put)

plt.plot(K_range, call_prices, label='Call')
plt.plot(K_range, put_prices, label='Put')
plt.xlabel('K')
plt.ylabel('Option Price')
plt.legend()
plt.show
```
